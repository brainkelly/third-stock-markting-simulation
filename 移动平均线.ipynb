```python
 def get_codes_by_date():
        bs.login()
        date = '2021-04-12'
        stock_rs = bs.query_all_stock(date)
        stock_df = stock_rs.get_data()
        stock_df = stock_df[stock_df.code>='sh.600000']
        stock_df = stock_df[stock_df.code<='sz.399000']
        stock_df = stock_df.reset_index()
        return stock_df
```

```python
def open_csv(path,row):
    dstdir = f'{path}/{row["code"]}.csv'
    df = pd.read_csv(dstdir)
    return df
```

```python
def df_process(df):
    df['new'] = df['close']
    df.new = df.new.shift(1)
    df['price_change'] = df.close - df.new 
    df['price_change'][0]=df['close'][0]-df['open'][0]
    del df['new']
    return df
```

```python
def rolling_mean(nums,df,columns):
    df[str(nums)+ columns] = df[i].rolling(nums).mean()
    #df[str(nums)+ columns] =  df[str(nums)+ columns].shift(1)
    return df
```

```python
c = ['open', 'high', 'low', 'close', 'volume', 'amount','price_change','turn']
```

```python
df = get_codes_by_date()
for index,row in df.iterrows():
    try:
        path = '/Users/senlinlidewo/pyprogram/dianping_zz/total_merge'
        df1 = open_csv(path,row)
        df2 = df_process(df1)
        for i in c:
            df3 = rolling_mean(15,df2,i)
        dstdir = f'{path}/{row["code"]}.csv'
        df3.to_csv(dstdir, index=False)
    except:
        continue
```
